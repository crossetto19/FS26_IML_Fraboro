#### Transform features

```python
"""
Transform the 5 input features of matrix X (x_i denoting the i-th component of a given row in X) 
into 21 new features phi(X) in the following manner:
5 linear features: phi_1(X) = x_1, phi_2(X) = x_2, phi_3(X) = x_3, phi_4(X) = x_4, phi_5(X) = x_5
5 quadratic features: phi_6(X) = x_1^2, phi_7(X) = x_2^2, phi_8(X) = x_3^2, phi_9(X) = x_4^2, phi_10(X) = x_5^2
5 exponential features: phi_11(X) = exp(x_1), phi_12(X) = exp(x_2), phi_13(X) = exp(x_3), phi_14(X) = exp(x_4), phi_15(X) = exp(x_5)
5 cosine features: phi_16(X) = cos(x_1), phi_17(X) = cos(x_2), phi_18(X) = cos(x_3), phi_19(X) = cos(x_4), phi_20(X) = cos(x_5)
1 constant feature: phi_21(X)=1

Parameters
----------
X: matrix of floats, dim = (700,5), inputs with 5 features

Compute
----------
X_transformed: matrix of floats: dim = (700,21), transformed input with 21 features
"""
X_transformed = np.zeros((700, 21))
# TODO: Enter your code here

for i in range(5):
    X_transformed[:,i] = X[:,i]
for i in range(5,10):
    X_transformed[:,i] = (X[:,i-5])**2
for i in range(10,15):
    X_transformed[:,i] = np.exp(X[:,i-10])
for i in range(15,20):
    X_transformed[:,i] = np.cos(X[:,i-15])

X_transformed[:,20] = 1



assert X_transformed.shape == (700, 21)
```

```python
"""
Alternative faster way proposed here:
"""

# 5 linear features
X_transformed[:, 0:5] = X

# 5 quadratic features
X_transformed[:, 5:10] = X**2

# 5 exponential features
X_transformed[:, 10:15] = np.exp(X)

# 5 cosine features
X_transformed[:, 15:20] = np.cos(X)

# 1 constant feature
X_transformed[:, 20] = 1

assert X_transformed.shape == (700, 21)
```

#### Fit data

```python
"""
Use the transformed data points X_transformed and fit the logistic regression on this 
transformed data. Finally, compute the weights of the fitted logistic regression. 

Parameters
----------
X_transformed: array of floats: dim = (700,21), transformed input with 21 features
y: array of integers \in {0,1}, dim = (700,), input labels

Compute
----------
w: array of floats: dim = (21,), optimal parameters of logistic regression

METHOD 1: with sci-kit learn Logistic Regression built in class
"""

weights_scikit = np.zeros((21,))
# TODO: Enter your code here

from sklearn.linear_model import LogisticRegression

# the last feature, phi21 is 1, which is equivalent to fitting an intercept, so it is set to false
LogReg_fit = LogisticRegression(fit_intercept=False).fit(X_transformed,y)

print(LogReg_fit.coef_[0])

weights_scikit = LogReg_fit.coef_[0]

assert weights_scikit.shape == (21,)
```

```python
# Save results in the required format
np.savetxt("./task1b_results_scikit_cr.csv", weights_scikit, fmt="%.12f")
```

```python
"""
Use the transformed data points X_transformed and fit the logistic regression on this 
transformed data. Finally, compute the weights of the fitted logistic regression. 

Parameters
----------
X_transformed: array of floats: dim = (700,21), transformed input with 21 features
y: array of integers \in {0,1}, dim = (700,), input labels

Compute
----------
w: array of floats: dim = (21,), optimal parameters of logistic regression

METHOD 2: manually with the Numpy package
"""
weights_manual = np.zeros((21,))
# TODO: Enter your code here

# defining the Sigmoid activation function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# setting the Gradient Descent hyperparameters
learning_rate = 0.1     # Also known as alpha, controls step size
num_iterations = 10000  # Number of times we update the weights
m = X_transformed.shape[0] # Number of training examples (700)

# gradient Descent Optimization Loop
for i in range(num_iterations):
    # Step A: Calculate the current predictions
    # h = sigmoid(X * weights)
    linear_predictions = np.dot(X_transformed, weights_manual)
    predictions = sigmoid(linear_predictions)
    
    # Step B: Calculate the gradient (how far off we are, scaled by X)
    # The derivative of the log-loss cost function simplifies beautifully to this:
    gradient = np.dot(X_transformed.T, (predictions - y)) / m
    
    # Step C: Update the weights by taking a step in the opposite direction of the gradient
    weights_manual -= learning_rate * gradient

assert weights_manual.shape == (21,)
```
